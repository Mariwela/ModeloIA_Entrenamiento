# 游끤 SUBTITULADOR AUTOM츼TICO DE VIDEOS OL칈MPICOS

# Hugging Face + Gradio para generar subt칤tulos de manera autom치tica a partir de un video.
# Modelo utilizado: openai/whisper-small

```python
from transformers import pipeline
import gradio as gr
from moviepy.editor import VideoFileClip
import tempfile, os

def subtitular_video(video_path):
    try:
        # Extraer audio temporal del video
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as temp_audio:
            clip = VideoFileClip(video_path)
            clip.audio.write_audiofile(temp_audio.name, codec="pcm_s16le")
            clip.close()
            audio_path = temp_audio.name

        # Cargar modelo desde Hugging Face (usamos Whisper debido a que soporta mas de 99 idiomas)
        asr = pipeline("automatic-speech-recognition", model="openai/whisper-small")

        # Generar transcripci칩n
        result = asr(audio_path)

        # Limpiar archivo temporal
        os.remove(audio_path)

        # Devolver el texto de los subt칤tulos
        return result["text"]

    except Exception as e:
        return f"丘멆잺 Error procesando el video: {str(e)}"

demo = gr.Interface(
    fn=subtitular_video,
    inputs=gr.Video(label="游꿘 Sube tu video ol칤mpico"),
    outputs=gr.Textbox(label="游닇 Subt칤tulos generados"),
    title="游끤 Subtitulador autom치tico de videos ol칤mpicos",
    description="Sube un video corto relacionado con los Juegos Ol칤mpicos y obt칠n sus subt칤tulos autom치ticos generados con Whisper (Hugging Face)."
)

# Lanzar la aplicaci칩n gradio
demo.launch(share=True, inline=False)
```
