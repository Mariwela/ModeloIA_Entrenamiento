```python
# CELDA 3
def extraer_texto_url(url):
    """
    Devuelve el texto limpio de una página web, evitando bloqueos con User-Agent.
    """
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/142.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()  # levanta error si status != 200
    soup = BeautifulSoup(response.text, "html.parser")

    # Tomar solo texto visible
    textos = soup.stripped_strings
    return " ".join(textos)
```

```python
# CELDA 4
def generar_resumen(prompt, modelo="gemini-2.5-flash"):
    """
    Genera texto usando Gemini a partir de un prompt.
    """
    response = client.models.generate_content(
        model=modelo,
        contents=prompt
    )
    return response.text
```

```python
# CELDA 5
url = "https://es.wikipedia.org/wiki/Juegos_ol%C3%ADmpicos"

# Extraer texto de la web
texto = extraer_texto_url(url)

# Crear prompt para resumir
prompt = f"Resume en 5 líneas la siguiente información:\n\n{texto[:2000]}..."  # limitar para no exceder token limit

# Generar resumen con Gemini
resumen = generar_resumen(prompt)
print(resumen)
```
